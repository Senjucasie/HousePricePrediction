```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv('melb_data.csv')
Data.head()
Y=Data.Price
X=Data.drop('Price',axis=1)
X.describe(include='all')
```

Remove categories with more number of Unique values

```python
maxuniquevalues=10
X.describe(include='all')
dropcol=[]
for colname in X:
   if(X[colname].nunique()>maxuniquevalues) and X[colname].dtype=='object':
    print(colname)
    dropcol.append(colname)
 
X=X.drop(dropcol,axis=1)
X_category=X.select_dtypes(include='object')
X_numerical=X.select_dtypes(include='number')



```

Splitting the data in to training and validation 

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,Y_train,Y_valid=train_test_split(X,Y,train_size=0.8,random_state=0)

#resetting the index of the data and assigning back to it
X_train.reset_index(drop=True,inplace=True)
X_valid.reset_index(drop=True,inplace=True)
X_train.describe(include='all')

```

Function to Calculate the Accuracy of the model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
def modelaccuracyscore(x_train,x_valid,y_train,y_valid):
    model=RandomForestRegressor(n_estimators=100,random_state=0)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_valid)
    print(mean_absolute_error(y_valid,y_pred)) 
    
```

Taking Care of Misssing numerical variables

```python
from sklearn.impute import SimpleImputer
imp_mean=SimpleImputer(missing_values=np.nan,strategy='mean')
num_x_train=X_train.select_dtypes(include='number')
num_x_valid=X_valid.select_dtypes(include='number')
num_column=num_x_train.columns

impute_x_train=pd.DataFrame(imp_mean.fit_transform(num_x_train))
impute_x_valid=pd.DataFrame(imp_mean.transform(num_x_valid))


impute_x_train.columns=impute_x_valid.columns=num_column
X_train=X_train.select_dtypes(include=object)
X_valid=X_valid.select_dtypes(include=object)
X_train=pd.concat([X_train,impute_x_train],axis=1)
X_valid=pd.concat([X_valid,impute_x_valid],axis=1)

    

        
  
```

Taking Care of Categorical values

Mehod 1 dropping coloumns with categorical variables

```python

x_train_removedc=X_train.select_dtypes(exclude=['object'])
x_valid_removedc=X_valid.select_dtypes(exclude=['object'])
modelaccuracyscore(x_train_removedc,x_valid_removedc,Y_train,Y_valid)
```

Method 2 Label encoding


```python
X_train_le=X_train.copy()
X_valid_le=X_valid.copy()
from sklearn.preprocessing import LabelEncoder
for col in X_category:
    X_train_le[col]=LabelEncoder().fit_transform(X_train[col])
    X_valid_le[col]=LabelEncoder().fit_transform(X_valid[col])

modelaccuracyscore(X_train_le,X_valid_le,Y_train,Y_valid)
```

Method 3 One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohencoder=OneHotEncoder(handle_unknown='ignore',sparse=False)
Oh_Xtrain=pd.DataFrame(ohencoder.fit_transform(X_train[X_category.columns]))
Oh_Xvalid=pd.DataFrame(ohencoder.fit_transform(X_valid[X_category.columns]))



num_xtrain=X_train.drop(X_category.columns,axis=1)
num_xvalid=X_valid.drop(X_category.columns,axis=1)
num_xtrain
Oh_X_t=pd.concat([num_xtrain,Oh_Xtrain],axis=1)
Oh_X_v=pd.concat([num_xvalid,Oh_Xvalid],axis=1)



modelaccuracyscore(Oh_X_t,Oh_X_v,Y_train,Y_valid)
```
